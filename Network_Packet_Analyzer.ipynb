```python
from scapy.all import sniff, IP, TCP, UDP, ICMP

# Define the callback function to process packets
def packet_callback(packet):
    # Print the packet summary
    print(f"Packet received: {packet.summary()}")

    # Extract and display source and destination IP addresses
    if IP in packet:
        src_ip = packet[IP].src
        dst_ip = packet[IP].dst
        print(f"Source IP: {src_ip}")
        print(f"Destination IP: {dst_ip}")

        # Extract and display the protocol
        if TCP in packet:
            print("Protocol: TCP")
        elif UDP in packet:
            print("Protocol: UDP")
        elif ICMP in packet:
            print("Protocol: ICMP")
        else:
            print("Protocol: Other")

        # Extract and display the payload data
        payload = packet.payload
        if payload:
            print(f"Payload Data (hex): {payload.original.hex()}")
            print(f"Payload Data (ascii): {payload.original.decode(errors='ignore')}")

# Define the main function to start sniffing
def main():
    print("Starting packet sniffer...")
    # Start sniffing packets, apply a filter to capture only IP packets
    sniff(prn=packet_callback, store=0, filter="ip", count=10)  # Set count=-1 for infinite sniffing

if __name__ == "__main__":
    main()
```
